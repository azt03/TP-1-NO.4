```python
#TP1 - NO.4 POIN.B
import numpy as np

# Atur jumlah lemparan
n_tosses = 2000

# Kemungkinan mendapatkan 1
p_heads = 0.55

# Simulasikan lemparan koin
tosses = np.random.binomial(1, p_heads, n_tosses)

# Hitung rata-rata kumulatif untuk 1s dan 0s
cumulative_avg_1s = np.cumsum(tosses) / np.arange(1, n_tosses + 1)
cumulative_avg_0s = 1 - cumulative_avg_1s

# Kembalikan rata-rata kumulatif untuk beberapa lemparan pertama dan nilai akhir
cumulative_avg_1s[:10], cumulative_avg_0s[:10], cumulative_avg_1s[-1], cumulative_avg_0s[-1]
```

```python
#TP1 - NO.4 POIN.C
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Menetapkan peluang kemunculan angka
p = 0.55

# Menyimulasikan 2000 lemparan koin
x = bernoulli.rvs(p, size=2000)

# Menghitung rata-rata kemunculan 1 setelah 1 hingga 2000 lemparan
averages_1 = np.cumsum(x) / np.arange(1, 2001)

# Menghitung rata-rata kemunculan 0 dengan cara 1 - rata-rata kemunculan 1
averages_0 = 1 - averages_1

# Membuat diagram garis
plt.figure(figsize=(12, 6))
plt.plot(range(1, 2001), averages_1, label='Rata-rata Kemunculan 1', color='blue')
plt.plot(range(1, 2001), averages_0, label='Rata-rata Kemunculan 0', color='red', linestyle='--')
plt.xlabel('Jumlah Lemparan')
plt.ylabel('Rata-rata Kemunculan')
plt.title('Perubahan Rata-rata Kemunculan 1 dan 0 setelah N Lemparan')
plt.legend()
plt.grid(True)
plt.show()
```
